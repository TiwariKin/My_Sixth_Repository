# Sales Data Analysis

```python
# Import neccesary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Import the dataset

data = pd.read_excel(r'C:\Users\Microsoft\Documents\dataset 1.xlsx')
data
```

- Get the number of rows and columns

```python
data.shape
```

- Count the number of missing values

```python
data.isnull().sum()
```

 - Which are the most selling products?

```python
boughts = data.groupby('Product').agg({'quantity': 'sum'}).sort_values('quantity',ascending=False)[:10]
boughts
```

```python
sns.barplot(x='Product',y='quantity',data=boughts.reset_index())
plt.xticks(rotation='vertical')
```

 - Which are the top 10 products by sales?

```python
ProductSales = pd.DataFrame(data.groupby('Product').sum()['Sales'])
ProductSales.sort_values(by=['Sales'],inplace=True,ascending=False)
ProductSales.head(20)
```

- Which are the most profitable products?

```python
Profitable_Products = pd.DataFrame(data.groupby(['Product']).sum()['profit'])
Profitable_Products.sort_values(by=['profit'],inplace=True,ascending=False)
Profitable_Products
```

```python
sns.barplot(x='Product',y='profit',data=Profitable_Products.head(15).reset_index())
plt.xticks(rotation = 'vertical')
```

- What category sold the most?

```python
data.groupby(['category','Sales']).sum()
```

- Which are the most profitable category?

```python
Profitable_Category = pd.DataFrame(data.groupby(['category']).sum()['profit'])
Profitable_Category.sort_values(by=['profit'],inplace=True,ascending=False)
Profitable_Category
```

```python
plt.figure(figsize=(17,5))
plt.bar(data['category'],data['Sales'],color='r')
plt.xticks(size=8)
plt.show()

```

- Total sales values by category and subcategory

```python
data.groupby(['category','sub_category']).sum()
```

- Which are the most selling products in subcategory?

```python
data.groupby(['category','sub_category'],as_index=False)['quantity'].count()
```

- Which customer segments are the most profitable?

```python
Profitable_Segment = pd.DataFrame(data.groupby(['segment']).sum()['profit'])
Profitable_Segment.sort_values(by=['profit'],inplace=True,ascending=False)
Profitable_Segment
```

```python
sns.barplot(x='segment',y='profit',data=Profitable_Segment.reset_index())
plt.xticks(rotation='vertical')
```

- What shipping modes sold the most products?

```python
plt.figure(figsize=(14,6))
plt.bar(data['Ship_mode'],data['Sales'],color='g')
plt.xticks(rotation='vertical',size=8)
plt.show()
```

- Visualize the 'Category' column from the shipmode column dataset standpoints. 

```python
category_hist = sns.FacetGrid(data, col='Ship_mode', palette='rainbow')
category_hist.map(plt.hist, 'category')
```

- What market sold the most products?

```python
sns.set_style('whitegrid')
sns.countplot(x='market',data=data, palette='rainbow')
```

- Which are the top 10 countries by sales?

```python
countries = pd.DataFrame(data.groupby('Country').sum()['Sales'])
countries.sort_values('Sales',inplace=True,ascending=False)
countries.head(15)
```
